```python
import pandas as pd
from pandas import Series, DataFrame
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
import os
import re
```

```python
dir_list = os.listdir('./downloadfile/')
```

```python
np.sort(dir_list)
```

```python
num_list_sort = np.sort([re.sub('data','',i) for i in dir_list])[:5]
dir_list_sort = np.sort(dir_list)[:5]
```

```python
num_list_sort
```

```python
p_d = []
a_d = []
```

```python
for i in range(len(dir_list_sort)):
    p_d.append(pd.read_csv('/home/wn/桌面/my_py_pro/downloadfile_script/downloadfile/'+dir_list_sort[i]+'/result'+num_list_sort[i]+'.csv'))
    a_d.append(pd.read_csv('/home/wn/桌面/my_py_pro/downloadfile_script/downloadfile/'+dir_list_sort[i]+'/input_data'+num_list_sort[i]+'.csv'))
```

```python
if not os.path.exists('./save_pic/'): os.mkdir('./save_pic/')
for i in range(len(p_d)):
    plt.figure(figsize=(50,10))
    ax1 = plt.subplot(211)
    ax1.set_title('predict')
    ax1.plot(p_d[i].iloc[:,-1])
    ax2=plt.subplot(212)
    ax2.set_title('active')
    ax2.plot(a_d[i].iloc[60:,-1])
    plt.tight_layout(2)
    plt.savefig('./save_pic/compare'+str(i)+'.png')
```
